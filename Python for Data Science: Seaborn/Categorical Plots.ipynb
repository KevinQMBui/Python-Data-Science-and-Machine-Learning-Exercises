```python
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Error Bar Video: https://www.youtube.com/watch?v=_7IUHwMNYG4
# If error bar (pointy line) overlaps, it's not significant

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
# bar plot displays mean values by default, so you can use the estimator
# function to pass sum, standard deviation, etc. instead
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips,estimator=np.std)
```

```python
# counts number of occurances as a bar plot. No need for y-axis
sns.countplot(x='sex',data=tips)
```

```python
# Box Plot & Skews: https://www.youtube.com/watch?v=L68iieC2Vgc
# Good way to view a distribution
# Tiny dots are outliers
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# https://www.youtube.com/watch?v=cLHwwRgny5g
# Ideal for observing an entire population rather than a select sample
# Similar to box plot, but with more information. Learn this.
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# split function allows us to compare different hue variables side-by-side
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
# jitter is True by default
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# 'split' parameter will be renamed to 'dodge'
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# Combination of a strip and violin plot
# Shows same distribution, but showing all the points themselves
# Do not use on large data sets. Not really used much
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# can be useful to combine the violin and swarm plot together
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# 'factorplot' will be renamed to 'catplot' in a future update
# more general way of calling different types of plots
# kind='bar, count, box, violin, strip, swarm, etc.' to choose plot type
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
