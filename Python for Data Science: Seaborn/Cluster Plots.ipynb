```python
import seaborn as sns
%matplotlib inline
sns.set()
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# you can find other markers by referencing some matplotlib stuff
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
# made the markers bigger
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
# rather than separate by color with 'hue', could use 'col' to separate
# by different columns/graphs instead. Same with 'row'
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
# great to change graph size to make it easier to see
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
           aspect=0.6,height=8)
```
