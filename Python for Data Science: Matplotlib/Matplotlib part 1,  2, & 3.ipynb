```python
# Figure Size and DPI
```

```python
# figsize=(width, height), dpi = dots per inch
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
ax.plot(x,x**4)

ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.plot(x,x**4,label='X Quad')

ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend(loc=0)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# can also use RGB text codes too instead of color names like 'orange'
# linewidth/lw=N changes size of line
# alpha=0-1 makes line more transparent
# linestyle/ls changes style of line. Can use, '--',':','dash', etc.
# marker options add and customizes marker
ax.plot(x,y,color='purple',lw=2,ls='--',marker='o',markersize=8,
       markerfacecolor='yellow',markeredgewidth=1,markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])


ax.plot(x,y,color='purple',lw=2,ls='--',marker='o',markersize=8,
       markerfacecolor='yellow',markeredgewidth=1,markeredgecolor='green')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
