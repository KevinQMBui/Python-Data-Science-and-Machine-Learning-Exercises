```python
# First unsupervised machine learning model I'm using!
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
# create blob-like groups of artificial data to make clusters
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,
                  cluster_std=1.8,random_state=101)
```

```python
data[0] # xy-coordinates of data plotted on a point
```

```python
data[1] # label for each of the plot points
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
data[1]
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='Accent')
# One cluster (orange) is distinct from the other three.
# The other three clusters are pretty noisy between each other.
```

This is where we actually start utilizing the KMeans model to cluster.

```python
from sklearn.cluster import KMeans
```

```python
# Must know the amount of clusters (shown in plot, so 4) beforehand
kmeans = KMeans(n_clusters=4)
```

```python
# Fit the KMeans model to the features. No need to test-train-split
kmeans.fit(data[0])
```

```python
# Center of each of the 4 clusters in xy coordinate
kmeans.cluster_centers_
```

```python
# If I was working with real data, I would be done at this stage
# Used to find labels in unlabled datasets
# However, we have the answers in data [1]
kmeans.labels_
```

```python
# Comparing original data plot with K Means predicted plot
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='Accent')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='Accent')
```
